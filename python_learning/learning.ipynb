# Variables: labels you can assign to values
- strings: series of characters
    - f-strings: strings prefixed with f that allow variables or expressions to be inserted inside {}
    - whitespace: any nonprinting character e.g. spaces, tabs or end-of-line symbols
        - \t -> add a tab
        - \n newline
        - stripping whitespace strip()
    - removing strings
        - removeprefix()
        - removesuffix()
